```python
places = ['ALTA', 'BRIGHTON', 'THAYNES', 'SNOWBIRD', 'PARLEY_SUMMIT', 'SUNDANCE']
def make_merged_df(places, mapping=column_mapping):
    df = pd.DataFrame()
    for place in places:
        tmp_df = (pd.read_csv('resources/weather/{}.csv'.format(place))
                    .rename(columns=mapping)
                    .add_prefix('{}_'.format(place))
                    .rename(columns={'{}_DATE'.format(place):'DATE'})
        )
        print(place, tmp_df.shape) 
        if df.empty:
            df = tmp_df
        else:
            df = pd.merge(df, tmp_df, on='DATE', how='outer')

    return df

df = make_merged_df(places)
df.shape

# Look at which datasets have too few of rows 
# Remove the datasets that have too few rows                        #remove Snowbird and Sundance
# Keep only as many rows as the dataset with the fewest rows        #keep only like 12908 rows
```
