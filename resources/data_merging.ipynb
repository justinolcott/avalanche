```python
import pandas as pd
```

```python
avy = pd.read_csv('resources/avalanche_report.csv')
```

```python
avy.head()
```

```python
weather = pd.read_csv('resources/weather/merged_weather.csv')
```

```python
weather.head()
```

```python
weather['DATE'].iloc[0]
```

```python
# make date columns of datetime type
# row 830 it switches format
avy['date'].iloc[:830] = pd.to_datetime(avy['date'].iloc[:830], format='%m/%d/%Y')
avy['date'].iloc[830:] = pd.to_datetime(avy['date'].iloc[830:], format='%Y-%m-%d')
```

```python
# rename date to DATE
avy.rename(columns={'date': 'DATE'}, inplace=True)
```

```python
# merge dataframes according to date
# for missing rows, just fill with nan
avy['DATE'] = pd.to_datetime(avy['DATE'])
weather['DATE'] = pd.to_datetime(weather['DATE'])
merged = pd.merge(avy, weather, on='DATE')
```

```python
for i, j in zip(merged.columns, merged.iloc[0]):
    print(i, ": ", j)
```

```python
# remove all columns that end in "_ATTRIBUTES"
cleaned_data = merged.loc[:, ~merged.columns.str.endswith('_ATTRIBUTES')]
```

```python
for i, j in zip(cleaned_data.columns, cleaned_data.iloc[0]):
    print(i, ": ", j)
```

```python
# sort by date
# redo the indexes
cleaned_data.sort_values(by=['DATE'], inplace=True)
cleaned_data.reset_index(drop=True, inplace=True)
```

```python
# save cleaned data
cleaned_data.to_csv('resources/DATA.csv', index=False)
```
